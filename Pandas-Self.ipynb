### Group by operations:

```python
import numpy as np
dt.info()
```

```python
dt.groupby('JOB_ID',as_index=False).agg({'SALARY':'mean'})
```

```python
dt.groupby('JOB_ID')[['SALARY','MANAGER_ID']].agg(['max','mean'])
```

```python
dt.groupby('JOB_ID',as_index=False)[['SALARY','MANAGER_ID']].agg({'SALARY':np.mean,'MANAGER_ID':np.count_nonzero})
```

```python
dt.groupby('JOB_ID',as_index=False).agg({'SALARY':'mean'}).rename(
columns={'SALARY':'AVG_SALARY'}).sort_values('AVG_SALARY',ascending=False)
```

### Adding NEW COLUMN :

```python
# Adding new columns
# Comparing Salary per Employee with average salary
dt['Hi_Low']=(dt['SALARY'])/(dt['SALARY'].mean())
```

```python
dt.head()
```

```python
dt['Sal_Dif']=(dt['SALARY']-dt['SALARY'].mean())
```

```python
dt
```

## `APPLY`

```python
#Apply
def get_sum(x):
    return np.sum(x)
dt[['SALARY','Sal_Dif']].apply(get_sum,axis=0)

#axis=0  : means compute row wise
#axis=1  :means compute column wise
```

## `Boolean Series`

```python
# Create a Boolean Series :
salary_flag=(dt['Hi_Low']*10 >5).map({True:'high',False:'low'})
print(salary_flag)
```
