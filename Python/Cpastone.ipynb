```python
import pandas as pd
import numpy as np
```

```python
# database urls
car_sales_transaction='Car_Sales_transactions.CSV'
city_url='City_Master.xlsx'
postcode_url='Postal_Code_Master.xlsx'
region_url='Region_Master.xlsx'
region_state_url='Region_State_Mapping.xlsx'
sales_status_url='Sales_Status.csv'
state_url='State_Master.xlsx'

```

```python
car_data=pd.read_csv(car_sales_transaction)
city_data=pd.read_excel(city_url)
postcode_data=pd.read_excel(postcode_url)
region_data=pd.read_excel(region_url)
region_state_data=pd.read_excel(region_state_url)
sales_status_data=pd.read_csv(sales_status_url)
state_data=pd.read_excel(state_url)
```

## merging all database

```python
#merging region_state_data with region_data
df_state_region=pd.merge(region_state_data,region_data,on='Region_Code',how='inner')
```

```python
df_state_region.head(3)
```

```python
#merging the above result with state data
df_state=pd.merge(df_state_region,state_data,on='State_Code',how='inner')
```

```python
df_state.head()
```

```python
# len(df_state['State_Code'].unique())
```

```python
#merge postcode_data with city_data on 'City_code'

df_city_postcode=pd.merge(city_data,postcode_data,on='City_Code',how='inner')
```

```python
df_city_postcode.head()
```

```python
# len(df_city_postcode['Postal_Code'].unique())
```

```python
# merging above result (df_city_postcode) with df_state on 'Sate_Code'
df_city_state=pd.merge(df_city_postcode,df_state,on='State_Code',how='inner')
```

```python
df_city_state
```

```python
#Merging the above result with car_data on 'Postal_Code'
# df_car_postcode=pd.merge(df_city_state,car_data,on='Postal_Code',how='inner')

df_car_postcode=pd.merge(df_city_state,car_data,how='inner',
                        left_on=['Postal_Code','City_Code','State_Code'],
                        right_on=['Postal_Code','City_Code','State_Code'])
```

```python
df_car_postcode.head()
```

```python
# len(df_car_postcode['Sales_ID'].unique())
```

```python
# merging the above result(df_car_postcode) with (sales_status_data) on 'Sales_ID'

final_data=pd.merge(df_car_postcode,sales_status_data,on='Sales_ID',how='inner')
```
